```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres du modèle
r = 0.05  # Taux de croissance intrinsèque
K = 100000  # Capacité limite (taille du marché)
P0 = 1000  # Nombre initial de clients

# Paramètres de simulation
delta_t = 0.01  # Essai avec un pas de temps plus petit
t_max = 100  # Temps maximum de simulation (en mois)
n_steps = int(t_max / delta_t)  # Nombre total d'étapes

# Initialisation des tableaux
t = np.zeros(n_steps + 1)
P = np.zeros(n_steps + 1)
P[0] = P0  # Condition initiale

# Implémentation de la méthode d'Euler
for n in range(n_steps):
    P[n+1] = P[n] + delta_t * r * P[n] * (1 - P[n]/K)
    t[n+1] = t[n] + delta_t
    
    # Pour trouver quand P atteint 50000
    if P[n] < 50000 and P[n+1] >= 50000:
        print(f"P atteint 50000 clients après environ {t[n+1]:.2f} mois")

# Affichage des résultats
plt.figure(figsize=(10, 6))
plt.plot(t, P)
plt.axhline(y=50000, color='r', linestyle='--', label='Objectif: 50000 clients')
plt.xlabel('Temps (mois)')
plt.ylabel('Nombre de clients')
plt.title('Croissance des clients d\'une startup SaaS')
plt.grid(True)
plt.legend()
plt.show()

# Afficher les premiers points et le résultat final
print(f"Clients initiaux: {P[0]}")
print(f"Clients après 12 mois: {P[int(12/delta_t)]:.0f}")
print(f"Clients après 24 mois: {P[int(24/delta_t)]:.0f}")
print(f"Clients après 60 mois: {P[int(60/delta_t)]:.0f}")
print(f"Clients finaux: {P[-1]:.0f}")
```

### Explications : 

Le premier calcul avec un pas de temps de 0,1 mois a donné une estimation de 92,10 mois pour atteindre 50 000 clients. La réduction du pas a affiné le résultat à 91,92 mois. Cette légère différence démontre l'impact du pas sur la précision du calcul. La valeur retenue de 91,92 mois (environ 7,7 ans) constitue l'estimation finale, avec la remarque que pour des décisions d'entreprise, l'ordre de grandeur reste plus significatif que la précision absolue (c'est à dire qu'en business, savoir que la croissance prendra environ 8 ans est plus utile que la précision à quelques jours près).
